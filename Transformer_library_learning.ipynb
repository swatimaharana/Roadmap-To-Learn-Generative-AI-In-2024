```python
reviews = ["I was dubious about the reviews of Pushpa2 after watching it","It was a worth a watch"]
results = classifier(reviews)
print(results)
```

```python
for review,result in zip(reviews,results):
  print(f"Review:{review} | Sentiment: {result['label']} | Score : {result['score']}")
```

```python
model = classifier.model
output = model(result['input_ids'],result['attention_mask'])
print(output)
```
